```python
#first step
f1=open('../itembased/outputs/mixed_2_cb.csv')
f2=open('../itembased/outputs/sub_item_based_J40_K10.csv')
f3=open('../itembased/outputs/mixed_cb_kNN_1042_2045_4050_32100_32384.csv','w+')
#mixed_cb_kNN_1042_2045_4050_32100_32384.csv
import csv

c1=csv.reader(f1)
c2=csv.reader(f2)
cw=csv.writer(f3)

l1=list(c1)
l2=list(c2)

cw.writerow(l1[0])
for i in range(1,4197):
    rec1=l1[i][1].split(' ')
    rec2=l2[i][1].split(' ')
    rec=""
    k=0
    for j in range (1,6):
        if int(rec1[j])<6:
            while k<6 and rec2[k] in rec1: 
                    k+=1
            if k<6: 
                rec=rec+" "+rec2[k]
                k+=1
            else: rec=rec+" "+rec1[j]
        else: rec=rec+" "+rec1[j]
    cw.writerow([l1[i][0],rec])
f3.close()
```
